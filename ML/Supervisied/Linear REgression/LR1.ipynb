```python
#import libarires
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importing Dataset
df=pd.read_csv("C:/Users\Sharath Reddy/OneDrive/Desktop/sharath/ML/supervisied/Linear REgression/income.data.csv")
df.head()
```

```python
#plot the Graph
plt.scatter(df.income,df.happiness,color="red",marker="*")
plt.xlabel("income")
plt.ylabel("happiness")
plt.show()
```

```python
X=df['income'].values.reshape(-1,1)
y=df['happiness'].values.reshape(-1,1)
```

```python
print(X)
```

```python
print(y)
```

```python
#spliting the dataset
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#importing the model
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
#fitting the model
lr.fit(X_train,y_train)
```

```python
#predict
y_pred=lr.predict(X_test)
y_pred
```

```python
print('valve of intercept:',lr.intercept_)
print('valve of coefficient:',lr.coef_)
```

```python
lr.coef_
```

```python
from sklearn import metrics
```

```python
rmse=np.sqrt(metrics.mean_squared_error(y_test,y_pred))
R2_score=metrics.r2_score(y_test,y_pred)
rmse
```
