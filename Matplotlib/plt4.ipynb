```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("data/crops.csv")
df
```

```python
plt.scatter(df.area,df.msp,label="msp")
plt.scatter(df.area,df.yields,)
plt.title("area vs msp")
plt.xlabel("area")
plt.ylabel("msp")
plt.legend()
plt.show()
```
