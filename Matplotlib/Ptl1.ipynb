```python
x =[1,2,3,4,5]
y =[5,7,9,3,2]
barplot =plt.bar(x,y)
for bar in barplot:
    yval =bar.get_height()
    print(yval)
    #plt.text(bar.get_x()+bar.get_width()/2.0,yval,int(yval),va="bottom")
    #plt.text(bar.get_x()+bar.get_width()/2.0,yval,int(yval,va="top")
    plt.text(bar.get_x()+bar.get_width()/2.0,yval,int(yval),va="center")
plt.title("sample")
plt.xlabel("students")
plt.ylabel("marks")
plt.show()
```

```python
x=["reddy","siva","teju","anu","abhi","sharath"]
y=[88,36,55,92,67,99]
c=["black","red","green","yellow","purple","blue"]
#c1=["red","red","red","green","green","green"]
y,x =zip(*sorted(zip(y,x)))
plt.bar(x,y, color=c,label="sorted data")
plt.legend()
plt.title("sample data")
plt.xlabel("students")
plt.ylabel("marks")
plt.show()
```
