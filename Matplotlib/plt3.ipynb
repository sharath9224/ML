```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
finaldf=pd.DataFrame({ 'cities':['delhi','mumbai','bnglr','hyd'],
                       '2017_rf':[200,300,150,250],
                       '2018_rf':[100,200,300,150],
                       '2019_rf':[150,250,400,100]})
finaldf
```

```python
fig=plt.figure()
ax1=fig.add_subplot(131)
finaldf.plot(x="cities",y='2017_rf',ax=ax1,kind="bar",title="2017rainfall")

ax2=fig.add_subplot(132)
finaldf.plot(x="cities",y='2018_rf',ax=ax2,kind="bar",title="2018rainfall")

ax3=fig.add_subplot(133)
finaldf.plot(x="cities",y='2019_rf',ax=ax3,kind="bar",title="2019rainfall")
plt.show()
```

```python
import numpy as np
```

```python
company=["google","amazon","tcs","wipro","infosys"]
revenue=[60,30,50,40,10]
profit=[26,36,45,12,15]
xpos=np.arange(len(company))
print(xpos)
```

```python
plt.bar(xpos,revenue,width=0.4,label="revenue")
plt.bar(xpos+0.4,profit,width=0.4,label="profit")
plt.title("revenue vs profit")
plt.xlabel("company")
plt.xticks(xpos,company)
plt.ylabel("revenue")
plt.legend()
plt.show()
```
